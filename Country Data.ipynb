https://www.kaggle.com/datasets/rohan0301/unsupervised-learning-on-country-data

**country**: Nom du pays

**child_mort**: Décès d'enfants de moins de 5 ans pour 1000 naissances vivantes

**exports**: Exportations de biens et services par habitant. Exprimé en % du PIB par habitant

**health**: Total des dépenses de santé par habitant. Exprimé en % du PIB par habitant

**imports**: Importations de biens et services par habitant. Exprimé en % du PIB par habitant

**income**: Revenu net par personne

**inflation**: La mesure du taux de croissance annuel du PIB total

**life_expec**: Le nombre moyen d'années qu'un nouveau-né vivrait si les schémas de mortalité actuels devaient rester les mêmes

**total_fer**: Le nombre d'enfants qui naîtraient de chaque femme si les taux actuels de fécondité par âge restaient les mêmes.

**gdpp**: Le PIB par habitant. Calculé comme le PIB total divisé par la population totale.

# PCA

```python
data
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.95, random_state=0)
df_st_pca = pd.DataFrame(pca.fit_transform(df_st))

np.cumsum(pca.explained_variance_ratio_)
```

# KMeans

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
```

```python
inertia = []
silhouette_coef = []
for i in range(2,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit_transform(df_st_pca)
    inertia.append(kmeans.inertia_)
    labels = kmeans.labels_
    silhouette_coef.append(silhouette_score(df_st_pca, labels, metric='euclidean')) 

plt.figure(figsize=(10,10))
plt.subplot(2,1,1)   
plt.plot(range(2,11),inertia, marker = "o")
plt.title('Elbow Method',fontsize=15)
plt.xlabel('Number of clusters',fontsize=10)
plt.ylabel('Sum of Squared distance',fontsize=10)

plt.subplot(2,1,2)
plt.plot(range(2,11), silhouette_coef)
plt.title('Silhouette Coef.',fontsize=15)
plt.xlabel('Number of clusters',fontsize=10)
plt.ylabel('Mean Silhouette Coefficient for all samples.',fontsize=10)
plt.show()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer, KElbowVisualizer
```

```python
count = 1
plt.figure(figsize=(20,20))
for i in range(2,7) :
    plt.subplot(5,2,count)
    '''
    Créer une instance KMeans pour un nombre différent de clusters
    '''
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=0)
    kmeans.fit_predict(df_st_pca)
    centroids = kmeans.cluster_centers_  
    labels = kmeans.labels_
    '''
    Créer une instance de SilhouetteVisualizer avec une instance de KMeans
    Monter le visualiseur
    '''
    plt.title(f'Silhouette Visualizer --> n_clusters = {i}')x
    s_visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick')
    s_visualizer.fit(df_st_pca)
    count+=1
    
    plt.subplot(5,2,count)
    sns.scatterplot(df_st_pca.values[:,0],df_st_pca.values[:,1], hue= labels, palette="deep")
    sns.scatterplot(centroids[:,0], centroids[:,1], s=100, color='red')


    count+=1
```

```python
plt.figure(figsize=(20,10))
for i in range(2,6) :
    plt.subplot(3,2,i+1)
    '''
    Créer une instance KMeans pour un nombre différent de clusters
    '''
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=0)
    '''
    Créer une instance de SilhouetteVisualizer avec une instance de KMeans
    Monter le visualiseur
    '''
    plt.title(f'K Elbow Vizualizer --> n_clusters = {i}')
    k_visualizer = KElbowVisualizer(kmeans, colors='yellowbrick')
    k_visualizer.fit(df_st_pca)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=100, random_state=0)
y = kmeans.fit_predict(df_st_pca)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(df_st_pca.values[:,0],df_st_pca.values[:,1], hue=labels, palette="deep")
sns.scatterplot(centroids[:,0], centroids[:,1], s=150, color = 'red')
plt.xlabel('life_expec')
plt.ylabel('child_mort')
plt.show()
```
